```python
#dependencies 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
clean_data_path = "../cleaned_data/Electric_Vehicle_Population_Size_History_By_County_Cleaned.csv"
```

```python
clean_full_data = pd.read_csv(clean_data_path)
```

```python
clean_full_data.head()
```

```python
# Change date format
clean_full_data["Date (formatted)"] = pd.to_datetime(clean_full_data["Date"],format="%B %d %Y")

# Group data by date, then summarise columns of interest
date_data_EV_count = clean_full_data.groupby("Date (formatted)")["Electric Vehicle (EV) Total"].sum()
date_data_total_count = clean_full_data.groupby("Date (formatted)")["Total Vehicles"].sum()
date_data_percent = (date_data_EV_count / date_data_total_count)*100
date_data_percent = date_data_percent.map("{:.2f}".format)

# Assemble columns into dataframe
date_data = pd.DataFrame(
    {"Total Electric Vehicle (EV)": date_data_EV_count,
     "Total Vehicles": date_data_total_count,
     "Percent Electric Vehicles": date_data_percent
    }
)

# Display data
date_data
```

```python
#creating new columns for just "years"
date_data["Year"] = date_data.index.year
date_data
```

```python
yearly_data = date_data.groupby('Year').sum()
yearly_data
                                
```

```python
#Yearly increase in electric vehilcle and total vehicles 
Percentage_change_data = yearly_data.pct_change()
Percentage_change_data
```

```python
#yearly percentage of electric vehicles column creation
percent_elec_veh = (yearly_data['Total Electric Vehicle (EV)']/ yearly_data['Total Vehicles']) * 100
percent_elec_veh
```

```python
yearly_data ["Share of EV "] = percent_elec_veh
yearly_data
```
